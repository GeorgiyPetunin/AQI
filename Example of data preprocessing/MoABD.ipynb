```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
```

```python
X = data.drop(['City'], axis = 1)
Y = data['City']

x_smote, y_smote = sm.fit_resample(X, Y)
```

```python
class_counts = data.City.value_counts()

plt.bar(class_counts.index, class_counts.values)

plt.title('Розподіл кількості екземплярів за містами до генерації')
plt.xlabel('Міста')
plt.ylabel('Кількість екземплярів')
plt.show()
```

```python
class_counts = y_smote.value_counts()

plt.bar(class_counts.index, class_counts.values)

plt.title('Розподіл кількості екземплярів за містами після генерації')
plt.xlabel('Міста')
plt.ylabel('Кількість екземплярів')
plt.show()
```

```python
df = pd.concat([y_smote, x_smote], axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
df
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

def metric(y_test, preds):
    mae = mean_absolute_error(y_test, preds)
    r2 = r2_score(y_test, preds)

    print(f'MAE = {mae}')
    print(f'r2 = {r2}')
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(['AQI', 'City'], axis=1)
y = df.AQI

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
from catboost import CatBoostRegressor

cb = CatBoostRegressor(random_state=42)

cb.fit(x_train, y_train)

cb_preds = cb.predict(x_test)

metric(y_test, cb_preds)

feature_importances_catboost = cb.get_feature_importance()
feature_importance_df_catboost = pd.DataFrame({'Feature': x.columns, 'Importance': feature_importances_catboost})
feature_importance_df_catboost = feature_importance_df_catboost.sort_values(by='Importance', ascending=False)

print(feature_importance_df_catboost)

plt.figure(figsize=(10, 6))
plt.bar(feature_importance_df_catboost['Feature'], feature_importance_df_catboost['Importance'])
plt.xticks(rotation=45, ha='right')
plt.title('Важливість ознак (CatBoost)')
plt.show()
```

```python
from xgboost import XGBRegressor

xg = XGBRegressor(objective="reg:linear", random_state=42)

xg.fit(x_train, y_train)

xg_preds = xg.predict(x_test)

metric(y_test, xg_preds)

feature_importances_xgboost = xg.feature_importances_
feature_importance_df_xgboost = pd.DataFrame({'Feature': x.columns, 'Importance': feature_importances_xgboost})
feature_importance_df_xgboost = feature_importance_df_xgboost.sort_values(by='Importance', ascending=False)

print(feature_importance_df_xgboost)

plt.figure(figsize=(10, 6))
plt.bar(feature_importance_df_xgboost['Feature'], feature_importance_df_xgboost['Importance'])
plt.xticks(rotation=45, ha='right')
plt.title('Важливість ознак (XGBoost)')
plt.show()
```
